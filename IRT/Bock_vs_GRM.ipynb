# GRM (Graded Response Model) vs Bock

This notebook illustrates the connection between Bock and homogenous GRM.

First version: 12/25/2021
This version: 1/3/2024

```python
import sys
sys.path.append("../")
import numpy as np
import matplotlib.pyplot as plt
from irt_models import (plot_figure,
                        slope_intercept_parameterization,
                        Bock_normalized_IRT_response,
                        GRM_cum,
                        GRM_response,
                        transform_GRM_To_Bock_IRT)
```

# 1: Example GRM

Example illustrating the use of GRM for 4 responses (thus 3 b values). Shows both the response function and the cumulative function

```python
a = 1
b = np.array([-5, 0, 5])
low = -10
high = 10
num = 100
theta = np.linspace(low, high, num)

y_cum = GRM_cum(a, b, theta)
y_resp = GRM_response(a, b, theta)

plot_figure(theta, (y_resp, y_cum), low, high, title =r"GRM",
            label=("0", "1", "2", "3", "1", "2", "3"),
            linewidth=(5, 5, 5, 5, 2, 2, 2))
```

# 2: Example GRM and corresponding Bock-IRT

Following two examples shows the GRM and an apparently matching Bock-IRT parameterization. The log plot then shows that those are actually only identical in approximation

```python
a = 1
# b = np.array([-5, 0, 5])
# b = np.array([-5, 1, 5])
b = np.array([-5, 2, 5])
low = -10
high = 10
num = 100
theta = np.linspace(low, high, num)

y_grm = GRM_response(a, b, theta)

a0, b0 = transform_GRM_To_Bock_IRT(a, b)
y_bock = Bock_normalized_IRT_response(a0, b0, theta)

plot_figure(theta, (y_grm, y_bock), low, high, title =r"GRM vs. Bock",
            label=("0", "1", "2", "3", "0", "1", "2", "3"),
            linewidth=(5, 5, 5, 5, 2, 2, 2, 2))

plot_figure(theta, (y_grm, y_bock), low, high, title =r"GRM vs. Bock",
            label=("0", "1", "2", "3", "0", "1", "2", "3"),
            linewidth=(5, 5, 5, 5, 2, 2, 2, 2),
            logscale=True, xmin=1e-8)
```
